```python
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
url_half = "https://github.com/mosolab/DIP/blob/master/assignments/Lab3_half100.png?raw=true"
url_chess ="https://github.com/mosolab/DIP/blob/master/assignments/Lab3_chess100.png?raw=true"


img_half = io.imread(url_half)
img_chess = io.imread(url_chess)


cv2_imshow( cv2.hconcat( (img_half,img_chess) ) ) 
```

```python
hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); 

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); 

plt.show();
```

```python
blur_half3 = cv2.blur(img_half, (3,3))
blur_half5 = cv2.blur(img_half, (5,5))
blur_half9 = cv2.blur(img_half, (9,9))
blur_half15 = cv2.blur(img_half, (15,15))
blur_half35 = cv2.blur(img_half, (35,35))

blur_chess3 = cv2.blur(img_chess, (3,3))
blur_chess5 = cv2.blur(img_chess, (5,5))
blur_chess9 = cv2.blur(img_chess, (9,9))
blur_chess15 = cv2.blur(img_chess, (15,15))
blur_chess35 = cv2.blur(img_chess, (35,35))
```

```python
hist_half3 = cv2.calcHist([blur_half3], [0], None, [256], [0, 256])
hist_half5 = cv2.calcHist([blur_half5], [0], None, [256], [0, 256])
hist_half9 = cv2.calcHist([blur_half9], [0], None, [256], [0, 256])
hist_half15 = cv2.calcHist([blur_half15], [0], None, [256], [0, 256])
hist_half35 = cv2.calcHist([blur_half35], [0], None, [256], [0, 256])

hist_chess3 = cv2.calcHist([blur_chess3], [0], None, [256], [0, 256])
hist_chess5 = cv2.calcHist([blur_chess5], [0], None, [256], [0, 256])
hist_chess9 = cv2.calcHist([blur_chess9], [0], None, [256], [0, 256])
hist_chess15 = cv2.calcHist([blur_chess15], [0], None, [256], [0, 256])
hist_chess35 = cv2.calcHist([blur_chess35], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(30,20))

plt.subplot(621)
plt.imshow(img_half, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(622)
plt.plot(hist_half), plt.xlim([0, 255]), plt.grid()

plt.subplot(623)
plt.imshow(blur_half3, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(624)
plt.plot(hist_half3), plt.xlim([0, 255]), plt.grid()

plt.subplot(625)
plt.imshow(blur_half5, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(626)
plt.plot(hist_half5), plt.xlim([0, 255]), plt.grid()

plt.subplot(627)
plt.imshow(blur_half9, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(628)
plt.plot(hist_half9); plt.xlim([0, 255]), plt.grid()

plt.subplot(629)
plt.imshow(blur_half15, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(6,2,10)
plt.plot(hist_half15), plt.xlim([0, 255]), plt.grid()

plt.subplot(6,2,11)
plt.imshow(blur_half35, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.subplot(6,2,12)
plt.plot(hist_half35); plt.xlim([0, 255]), plt.grid()
```
