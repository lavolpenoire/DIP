```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
start = 0.0
stop = 2.0
step = 0.01
t = np.arange(start, stop, step)
```

```python
s = np.sin(2 * np.pi * t)
```

```python
f1 = 1
f2 = 2
f3 = 3
f4 = 4
```

```python
s1 = np.sin(2 * np.pi * t * f1)
s2 = np.sin(2 * np.pi * t * f2)
s3 = np.sin(2 * np.pi * t * f3)
s4 = np.sin(2 * np.pi * t * f4)
```

```python
fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, s1, c='r',linewidth=4.0)
ax1.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція 1 ')
ax1.grid()

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(t, s2, c='g',linewidth=4.0)
ax2.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція 2 ')
ax2.grid()

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, s3, c='b',linewidth=4.0)
ax3.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція 3 ')
ax3.grid()

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(t, s4, c='y',linewidth=4.0)
ax4.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція 4 ')
ax4.grid()

plt.show()
```
